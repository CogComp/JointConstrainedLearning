```python
import json
task = "temporal"
if task == "temporal":
    instance_01 = {"sent_1": "Jacob called the police.", \
                   "e1_start_char": 6, \
                   "sent_2": "Bob hit Jacob.", \
                   "e2_start_char": 4}
    instance_02 = {"sent_1": "Bob hit Jacob.", \
                   "e1_start_char": 4, \
                   "sent_2": "Jacob called the police.", \
                   "e2_start_char": 6}
    instance_03 = {"sent_1": "I ate breakfast.", \
                   "e1_start_char": 2, \
                   "sent_2": "I ate lunch.", \
                   "e2_start_char": 2}
    instance_04 = {"sent_1": "I ate lunch.", \
                   "e1_start_char": 2, \
                   "sent_2": "I ate breakfast.", \
                   "e2_start_char": 2}
    input_list = [instance_01, instance_02, instance_03, instance_04]
    with open("temporal_example_input.json", 'w') as f:
        json.dump(input_list, f)
    # the model would tell you: 
    # instance_01: Vague(Before > After); 
    # instance_02: Before; 
    # instance_03: Before; 
    # instance_04: Vague (Before < After)
else:
    instance_01 = {"sent_1": "The explosion happened on Tuesday.", \
                   "e1_start_char": 4, \
                   "sent_2": "Victims were sent to the hospital.", \
                   "e2_start_char": 13}
    instance_02 = {"sent_1": "Victims were sent to the hospital.", \
                   "e1_start_char": 13, \
                   "sent_2": "The explosion happened on Tuesday.", \
                   "e2_start_char": 4}
    instance_03 = {"sent_1": "I ate breakfast.", \
                   "e1_start_char": 2, \
                   "sent_2": "I ate lunch.", \
                   "e2_start_char": 2}
    instance_04 = {"sent_1": "Sam goes to the school and studies Math.", \
                   "e1_start_char": 4, \
                   "sent_2": "Sam goes to the school and studies Math.", \
                   "e2_start_char": 27}
    input_list = [instance_01, instance_02, instance_03, instance_04]
    with open("subevent_example_input.json", 'w') as f:
        json.dump(input_list, f)
    # HiEve model would tell you:
    # instance_01: Parent-Child; 
    # instance_02: Child-Parent; 
    # instance_03: Coref; 
    # instance_04: NoRel

    # Joint model would tell you:
    # instance_01: Parent-Child; 
    # instance_02: NoRel; 
    # instance_03: Coref; 
    # instance_04: Parent-Child
```
